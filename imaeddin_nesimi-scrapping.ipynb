```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://az.wikisource.org/wiki/M%C3%BC%C9%99llif:%C4%B0mad%C9%99ddin_N%C9%99simi"
```

```python
def get_soup(url):
        
    
    response=requests.get(url)
            
           
    
    html_ins=response.content

    soup=bs(html_ins, 'html.parser')
    
    return soup
```

```python
soup = get_soup(url)
```

```python
menu = soup.find("div",{"class":"mw-body-content"})
```

```python
seirler = {}
# ul = menu.find('ul')
li_tags = menu.find_all("li")

for li_tag in li_tags:

    a_tags = li_tag.find_all("a")
    
    for a_tag in a_tags:

        seir_url = 'https://az.wikisource.org'+a_tag['href']
        seir_name = a_tag.text

        seirler[seir_name] = seir_url
```

```python
new_dict = {}
for i in list(seirler.keys())[3:]:
    new_dict[i] = seirler[i]
```

```python
new_dict
```

```python
def get_the_poem(url):
    
    soup=get_soup(url)
    
    
    poem = soup.find('div',{"class":"poem"})
    
    
    try:
        poem = poem.find("p").text
    except:
        poem= None

    
    return poem
```

```python
data = []
```

```python
for key in new_dict.keys():
    poem = get_the_poem(new_dict[key])
    poem = poem.strip('\n').strip().replace("\n\n\n","\n").replace("\n\n",'\n').replace("\n\n\n\n","\n")
    poem_ar = poem.split("\n")
    data.extend(poem_ar)
    
```
